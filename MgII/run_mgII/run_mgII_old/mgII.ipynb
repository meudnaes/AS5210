Some relevant imports and `helita` for output analysis.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from helita.sim import rh15d
from helita.vis import rh15d_vis
from IPython.display import HTML, display, set_matplotlib_formats

set_matplotlib_formats('svg')
plt.rc('legend', frameon=False)
plt.rc('figure', figsize=(7, 7 / 1.75)) # Larger figure sizes
plt.rc('font', size=12)
```

After loading the output into data, we look at the intensity in the region of 279 - 280.5 nm, where the Mg II lines are located.

```python
data = rh15d.Rh15dout('output/')
data.close()
wave = data.ray.wavelength
indices = np.arange(len(wave))[(wave > 279) & (wave < 280.5)]
```

At first the lines have 418 wavelength points.
After changing `WAVETABLE = ../../rh_ast5210/Atoms/wave_files/IRIS_NUV_full.wave` in `keyword.input`, the lines have 821 wavelength points.

### Investigating the line

```python
intensity = data.ray.intensity
fig, ax = plt.subplots()
ax.plot(wave[indices], intensity[0,0,indices])
ax.set_xlabel("Wavelength [nm]")
ax.set_ylabel("Intensity [W/(Hz nm2 sr)]")
ax.set_title("Mg II h & k");
```

```python
fig, ax = plt.subplots()
ax.plot(wave[:], intensity[0,0,:])
ax.set_xlabel("Wavelength [nm]")
ax.set_ylabel("Intensity [W/(Hz nm2 sr)]")
ax.set_title("All wavelength points");
```

### Investigating wavelength files

We read an existing an existing wavelength file with the function 
`helita.sim.rh15d.read_wave_file`

```python
IRIS_wave = rh15d.read_wave_file('../../rh_ast5210/Atoms/wave_files/IRIS_NUV_full.wave')
```

We investigate the content of the IRIS wavelength file

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(IRIS_wave, np.arange(0, len(IRIS_wave)))
ax.set_xlabel("Wavelength [nm]")
ax.set_ylabel("Index")
ax.set_title("IRIS wavelength file");
```

The wavelength file is mostly linear, but makes small jumps at 279.6 and 280.3.

We write a wavelength file, with the function `helita.sim.rh15d.make_wave_file`:

```python
# this will write wavelenghts from 650 to 660 nm, 0.01 nm spacing
rh15d.make_wave_file('my.wave', 650, 660, 0.01)
```

```python
rh15d.make_wave_file('new.wave', IRIS_wave[0], IRIS_wave[-1], 0.001)
my_wave = rh15d.read_wave_file('new.wave')
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(my_wave, np.arange(0, len(my_wave)))
ax.set_xlabel("Wavelength [nm]")
ax.set_ylabel("Index")
ax.set_title("My wavelength file");
```
